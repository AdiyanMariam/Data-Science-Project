```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
churn_rate = 100 * yes_churn / (no_churn+yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

```python
plt.figure(figsize=(10,4),dpi=200)
churn_rate.iloc[0].plot()
plt.ylabel('Churn Percentage');
```

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return "Over 48 Months"
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df.head(10)[['tenure','Tenure Cohort']]
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort', linewidth=0.5,alpha=0.5)
```

```python
X = df.drop(['Churn','customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report
```

```python
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(dt,X_test,y_test)
```

```python
imp_feats = pd.DataFrame(data=dt.feature_importances_,index=X.columns,columns=['Feature Importance']).sort_values("Feature Importance")
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats.sort_values('Feature Importance'),x=imp_feats.sort_values('Feature Importance').index,y='Feature Importance')
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
preds = rf.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(dt,X_test,y_test)
```
